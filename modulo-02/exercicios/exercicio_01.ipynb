# Exercicios

1. Criar as seguintes tabelas:

**Clientes**
- id_cliente: INT, PK
- nome: VARCHAR(50)
- email: VARCHAR(50)
- telefone: INT
- endereco: VARCHAR(80)
- cpf: INT

**Categorias**
- id_categoria: INT, PK
- nome: VARCHAR(50)

**Produtos**
- id_produto: INT, PK
- nome: VARCHAR(50)
- descricao: TEXT
- preco: REAL
- id_categoria: INT

**Pedidos**
- id_pedido: INT, PK
- id_cliente: INT, FK
- data_pedido: DATE
- status: VARCHAR(10)

**ItensPedido**
- id_pedido: INT, PK, FK
- id_item_pedido: INT, PK
- id_produto: INT, FK
- quantidade: INT
- preco_unitario: FLOAT


![](image_01.png)


```python
%%sql
BEGIN;
CREATE SCHEMA exercicios_01;
SET search_path TO exercicios_01;
DROP TABLE IF EXISTS exercicios_01.clientes;
CREATE TABLE IF NOT EXISTS exercicios_01.clientes
(
    id_cliente serial,
    nome character varying(50) NOT NULL,
    email character varying(50) NOT NULL,
    telefone bigint NOT NULL,
    cpf bigint NOT NULL,
    PRIMARY KEY (id_cliente),
    CONSTRAINT cpf_unique UNIQUE (cpf)
);

DROP TABLE IF EXISTS exercicios_01.pedidos;
CREATE TABLE IF NOT EXISTS exercicios_01.pedidos
(
    id_pedido bigserial,
    id_cliente bigserial NOT NULL,
    data_pedido date NOT NULL,
    status character varying(10) NOT NULL,
    PRIMARY KEY (id_pedido)
);

DROP TABLE IF EXISTS exercicios_01.itens_pedido;
CREATE TABLE IF NOT EXISTS exercicios_01.itens_pedido
(
    id_item_pedido bigserial,
    id_pedido bigserial,
    quantidade integer NOT NULL,
    preco_unitario real NOT NULL,
    id_produto bigserial NOT NULL,
    PRIMARY KEY (id_item_pedido, id_pedido)
);

DROP TABLE IF EXISTS exercicios_01.categorias;
CREATE TABLE IF NOT EXISTS exercicios_01.categorias
(
    id_categoria bigserial,
    nome character varying(50) NOT NULL,
    PRIMARY KEY (id_categoria)
);

DROP TABLE IF EXISTS exercicios_01.produtos;
CREATE TABLE IF NOT EXISTS exercicios_01.produtos
(
    id_produto bigserial,
    nome character varying(50) NOT NULL,
    descricao text,
    preco real NOT NULL,
    id_categoria bigserial,
    PRIMARY KEY (id_produto)
);

ALTER TABLE IF EXISTS exercicios_01.pedidos
    ADD CONSTRAINT id_cliente FOREIGN KEY (id_cliente)
    REFERENCES exercicios_01.clientes (id_cliente) MATCH SIMPLE
    ON UPDATE NO ACTION
    ON DELETE NO ACTION
    NOT VALID;

ALTER TABLE IF EXISTS exercicios_01.itens_pedido
    ADD FOREIGN KEY (id_pedido)
    REFERENCES exercicios_01.pedidos (id_pedido) MATCH SIMPLE
    ON UPDATE NO ACTION
    ON DELETE NO ACTION
    NOT VALID;

ALTER TABLE IF EXISTS exercicios_01.itens_pedido
    ADD FOREIGN KEY (id_produto)
    REFERENCES exercicios_01.produtos (id_produto) MATCH SIMPLE
    ON UPDATE NO ACTION
    ON DELETE NO ACTION
    NOT VALID;

ALTER TABLE IF EXISTS exercicios_01.produtos
    ADD FOREIGN KEY (id_categoria)
    REFERENCES exercicios_01.categorias (id_categoria) MATCH SIMPLE
    ON UPDATE NO ACTION
    ON DELETE NO ACTION
    NOT VALID;

END;
```
