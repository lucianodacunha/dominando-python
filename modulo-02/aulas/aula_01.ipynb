# O que é Banco de Dados?

- O que é um dado?
- Como transformar dado em conhecimento?
- Como organizar os dados para a tomada de decisão?

## O que é dado?

- nome, sobrenome, cpf, telefone, etc.
- Dado é a matéria-prima, fato bruto sem contexto e sem interpretação
- Informação é o resultado da organização e interpretação dos dados

**Pirâmide DIKW**

- Sabedoria
- \+ insights
- Conhecimento
- \+ significado
- Informação
- \+ contexto
- Dado

## Banco de dados

Conjunto organizado de dados que pode ser acessado, gerenciado e atualizado

- Representa aspectos da vida real
- Podem ser armazenados de diversas formas
- Podem ser de qualquer tamanho
- SGBD: Sistema responsável por gerenciar o BD

**Importância**

- Organização e eficiência
- Tomada de decisão baseada em dados
- Escalabilidade

**Tipos de Bancos de Dados**

- SQL: Relacional
- NOSQL: Colunar, Vector DB, Documentos, Chave-Valor

**Como funcionam os bancos de dados?**

- Usuários final, através de um aplicação como interface, acessa o banco de dados.
- A aplicação, através de drives de conexão acessa a base
- A base de dados é composta por tabelas
- Tabelas, devem respeitar algumas regras de integridade dos dados
- Chaves primárias
- Chaves estrangeiras
- Outras constraints

## SQL - Structured Query Language

**DDL**: 

- create, alter, drop, truncate, rename

**DML**: 

- select, insert, update, delete

**DCL**:

- grant, revoke

**TCL**:

- commit, rollback, saveprint

### Modelagem de dados

- Modelagem Conceitual: entendimento do negócio, DER
- Modelagem Lógica: Traduão do modelo conceitual para relacional
- Modelagem Física: utilização de soluções para implementação real no SGBD, 
exemplo SQLDBM.com.
